```python
import cv2
import numpy as np
import datetime
import time
```

```python
def get_timestamp():
    timestamp=str(datetime.datetime.now())
    stamps=timestamp.split(" ")
    date=stamps[0]
    time=stamps[1]
    minutes=time.split(":")[1]
    timestamp=timestamp[:-7]
    return timestamp,date,time
timestamp,date,time=get_timestamp()
print(timestamp)
#print(time)
```

```python
def logData(filename,timestamp,new_face):
    print("logging")
    file1=open(filename,"a")
    data=timestamp+","+str(new_face)+"\n"
    file1.write(data)

cap=cv2.VideoCapture(0)
frame_width=int(cap.get(3))
frame_height=int(cap.get(4))

face_file="haarcascade_frontalface_default.xml"
face=cv2.CascadeClassifier(face_file)

def detect_face(input_array,face):
    gray=cv2.cvtColor(input_array,cv2.COLOR_RGB2GRAY)
    f1=face.detectMultiScale(gray,1.15,6)
    copied=input_array.copy()
    f_cnt=0
    for(x,y,w,h) in f1:
        f_cnt+=1
        cv2.rectangle(copied,(x,y),(x+w,y+h),(0,0,255))
    print(str(f_cnt)+"faces detected")
    return copied,f_cnt
no_of_persons=0
if cap.isOpened() is True:
    cnt=0
    while(True):
        ret,frame=cap.read()
        cnt+=1
        frame=cv2.flip(frame,1)
        path="ani"+str(cnt)+".jpeg"
        
        output,count=detect_face(frame,face)
        if(count>0):
            if(no_of_persons!=count):
                timestamp,date,time= get_timestamp()
                logData("logger"+date+".txt",timestamp,count)
        no_of_persons=count
        
        cv2.imwrite(path,output)
        cv2.imshow("Face-Detected",output)
        
        if cv2.waitKey(2)==27:
            break
cap.release()
cv2.destroyAllWindows()
```
